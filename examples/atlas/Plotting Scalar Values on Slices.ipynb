# Plotting scalar values on histology slices

This example walks through various ways to plot scalar brain region values on histology slices

```python
from ibllib.atlas.plots import plot_scalar_on_slice
from ibllib.atlas import AllenAtlas
import numpy as np

ba = AllenAtlas()
acronyms = np.array(['VPM', 'VPL', 'PO', 'LP', 'CA1', 'DG-mo'])
values = np.random.randn(len(acronyms))
```

## Example 1: Coronal slice with image background

```python
from ibllib.atlas.plots import plot_scalar_on_slice

fig, ax = plot_scalar_on_slice(acronyms, values, coord=-2000, slice='coronal', mapping='Allen', hemisphere='both', 
                               background='image', cmap='Reds', brain_atlas=ba)
```

## Example 2: Sagittal slice with boundary background

```python
from ibllib.atlas.plots import plot_scalar_on_slice

fig, ax = plot_scalar_on_slice(acronyms, values, coord=-2000, slice='sagittal', mapping='Allen', hemisphere='both', 
                               background='boundary', cmap='Reds', brain_atlas=ba)
```

## Example 3: Plot on flatmap

```python
from ibllib.atlas.plots import plot_scalar_on_flatmap
acronyms = np.unique(ba.regions.acronym)
values = np.random.randn(len(acronyms))
values[0] = 0
fig, ax = plot_scalar_on_flatmap(acronyms, values, depth=0, hemisphere='both', cmap='Reds')
```
